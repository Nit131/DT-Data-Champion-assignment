```python
import time
import pandas as pd
from bs4 import BeautifulSoup
import requests
from datetime import timedelta
from statistics import mode

start_time = time.time()
dataframe_final = pd.DataFrame()

total_number_of_webpages = 200

for page in range(1, total_number_of_webpages + 1):
    print("Scraping webpage number: {page} of {total}".format(page=page, total=total_number_of_webpages))
    loop_time = time.time()
    url = f"https://www.vcsdata.com/companies-hyderabad.php?page={page}&city=Hyderabad"
    header = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36"
    }
    response = requests.get(url, headers=header)
    soup = BeautifulSoup(response.text, 'lxml')
    company_cards = soup.find_all("div", class_="d-block w-100")
    name = []
    city = []
    postal_code = []
    country = []
    address = []
    industry = []
    for card in company_cards:
        name.append(card.find("h3").text.strip() if card.find("h3") else "N/A")
        industries = card.find_all("div")[1].text.strip() if len(card.find_all("div")) > 1 else "N/A"
        industry.append(industries)
        address.append(card.find("span", itemprop="streetAddress").text.strip() if card.find("span", itemprop="streetAddress") else "N/A")
        city.append(card.find("span", itemprop="addressLocality").text.strip() if card.find("span", itemprop="addressLocality") else "Hyderabad")
        postal_code.append(card.find("span", itemprop="postalCode").text.strip() if card.find("span", itemprop="postalCode") else "000000")
        country.append(card.find("span", itemprop="addressCountry").text.strip() if card.find("span", itemprop="addressCountry") else "India")
    col_dic = {
        "Company Name": name,
        "City": city,
        "Postal Code": postal_code,
        "Address": address,
        "Industry": industry,
        "Country": country
    }
    df = pd.DataFrame(col_dic)
    dataframe_final = pd.concat([dataframe_final, df], ignore_index=True)
    print("Page scraped successfully!")
    print("Time taken:", round((time.time() - loop_time) * 1000, 2), "ms")
    print("Total time elapsed:", str(timedelta(seconds=(time.time() - start_time))))
    print()

valid_postal_codes = [pc for pc in dataframe_final["Postal Code"] if pc != "000000"]
postal_code_mode = mode(valid_postal_codes) if valid_postal_codes else "000000"
dataframe_final["Postal Code"] = dataframe_final["Postal Code"].replace("000000", postal_code_mode)

end_time = time.time()
print("Full website scraped successfully!")
print("Total time taken:", str(timedelta(seconds=(end_time - start_time))))
```
